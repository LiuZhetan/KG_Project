```python
import json

import pandas as pd
import requests

# 试题知识点推荐功能测试
url = 'http://localhost:5000/question/'
params = {
    'question': '2．已知：①Al(OH)3固体的熔点为300 ℃，电离方程式为H＋＋AlO－2＋H2O==Al(OH)3==3OH－＋Al3＋；②无水AlCl3晶体易升华，溶于水的电离方程式为AlCl3===Al3＋＋3Cl－；③熔融状态的HgCl2不能导电，稀溶液具有弱的导电能力且可作为手术刀的消毒液。下列关于Al(OH)3、AlCl3和HgCl2的说法正确的是(　　) A．均为强电解质  	B．均为弱电解质 C．均为离子化合物  	D．均为共价化合物',
    'subject': '化学',
    'row_limit': '5'
}

response = requests.get(url, params=params)
print(response.text)
# 解析json文本
response = json.loads(response.text)
if response["return_result"]:
    result = response["result"]
    for level, score_json in result.items():
        print(f"knowledge level:{level}")
        print(f"knowledge score:\n{pd.DataFrame(json.loads(score_json))}")
else:
    print("not get result")
```

```python
#接下来对知识图谱的管理模块接口进行测试

# 根据id查询
url = 'http://localhost:5000/query_by_id/'
params = {
    'type': 'knowledge',
    'id': '2',
}

response = requests.get(url, params=params)
print("raw response:")
print(response.text)
# 解析json文本
response = json.loads(response.text)

if response["return_result"]:
    result = response["result"]
    print(f"query result:\n{pd.DataFrame(json.loads(result))}")
else:
    print("not get result")
```
